```python
#搜尋出股票特定的年份
import os
from datetime import datetime
import pandas as pd
current = os.path.abspath("./")
current_date = datetime.now()
stockid = "2303.TW"
filename = f"{stockid}_{current_date.year}_{current_date.month}_{current_date.day}.csv"
file_path = os.path.join(current,'data',filename)
if not os.path.exists(file_path):
    print("建立檔案")
else:
    print("讀取檔案")

year = "2019"
starYear = f'{year}-1-1'
stopYear = f'{year}-12-31'
stock_dataFrame = pd.read_csv(file_path)
stock_dataFrame['Date'] = pd.to_datetime(stock_dataFrame["Date"])
stock_dataFrame
#stock_dataFrame.info()

mask = (stock_dataFrame["Date"] >= starYear) & (stock_dataFrame["Date"] <= stopYear)

year_stock_dataFrame = stock_dataFrame.loc[mask]
year_stock_dataFrame
```

```python
import plotly.express as px
fig = px.line(year_stock_dataFrame,x="Date",y="Adj Close",title=f'{stockid}_{year}')
fig.show()
```
